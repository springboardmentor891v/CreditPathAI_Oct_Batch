```python
# Drop duplicates
df = df.drop_duplicates()

# Rename columns for clarity (slightly different from original notebook)
df.rename(columns={
    "Credit_Score": "CreditScore",
    "Annual_Income": "Income",
    "Monthly_Inhand_Salary": "Salary"
}, inplace=True)

df.head()
```

```python
df.isnull().sum()
```

```python
# Fill missing numeric values with median
num_cols = df.select_dtypes(include=[np.number]).columns
for col in num_cols:
    df[col] = df[col].fillna(df[col].median())

# Fill missing categorical values with mode
cat_cols = df.select_dtypes(include=['object']).columns
for col in cat_cols:
    df[col] = df[col].fillna(df[col].mode()[0])

df.isnull().sum()
```

```python
df['Default'] = df['Credit_Worthiness'].apply(lambda x: 1 if x == 'Bad' else 0)
df[['Credit_Worthiness', 'Default']].head()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x=df['Status'])
plt.title("Count of Default vs Non-Default")
plt.xlabel("Status (0=No, 1=Yes)")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='Status', y='income', data=df)
plt.title("Income vs Default")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(data=df, x='CreditScore', hue='Status', kde=False, bins=30)
plt.title("Credit Score Distribution by Default Status")
plt.xlabel("Credit Score")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='Status', y='age', data=df)
plt.title("Age vs Default")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(data=df, x='loan_amount', hue='Status', bins=30)
plt.title("Loan Amount Distribution by Default Status")
plt.xlabel("Loan Amount")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.kdeplot(data=df, x='income', hue='Status', fill=True)
plt.title("Income Distribution by Default Status")
plt.xlabel("Income")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.kdeplot(data=df.dropna(subset=['rate_of_interest']), 
            x='rate_of_interest', 
            hue='Status', 
            fill=True)
plt.title("Interest Rate Distribution by Default Status")
plt.xlabel("Interest Rate (%)")
plt.show()
```

```python
sns.boxplot(data=df, x='Default', y='open_credit')
plt.title('Open Credit Lines vs Default')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

categorical_cols = [
    'Gender', 'approv_in_adv', 'loan_type', 'loan_purpose',
    'Credit_Worthiness', 'business_or_commercial', 'interest_only',
    'construction_type', 'occupancy_type', 'Secured_by',
    'credit_type', 'co-applicant_credit_type',
    'submission_of_application', 'Region', 'Security_Type', 'Status'
]

for col in categorical_cols:
    plt.figure(figsize=(8,5))
    sns.countplot(data=df, x=col, hue='Default')
    plt.title(f'{col} vs Default')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.xticks(rotation=30)
    plt.tight_layout()
    plt.show()
```
