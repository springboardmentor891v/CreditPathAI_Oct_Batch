```python
import matplotlib.pyplot as plt
import seaborn as sns
import math

cols = df.columns
n = len(cols)

cols_per_row = 3
rows = math.ceil(n / cols_per_row)


fig, axes = plt.subplots(rows, cols_per_row, figsize=(cols_per_row * 5, rows * 4))
axes = axes.flatten()

for i, col in enumerate(cols):
    sns.boxplot(x=df[col], ax=axes[i], color='skyblue')
    axes[i].set_title(f'Boxplot for {col}')
    axes[i].set_xlabel(col)


for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
#and also we dont't have outliers
```

```python
# Set visualization style
sns.set(style="whitegrid", palette="viridis")
plt.rcParams["figure.figsize"] = (10, 6)
```

```python
#Target Variable Analysis
plt.figure(figsize=(6,4))
sns.countplot(x='Default', data=df, palette='Set2')
plt.title("Loan Default Distribution")
plt.xlabel("Default (1 = Defaulted, 0 = Not Defaulted)")
plt.ylabel("Count")
plt.show()

print("\nDefault Rate: {:.2f}%".format(df['Default'].mean() * 100))
```

```python
#Numeric Features Exploration

numeric_cols = df.select_dtypes(include=['int64','float64']).columns.drop(['Default'])

df[numeric_cols].hist(bins=30, figsize=(15, 10), layout=(3, 3))
plt.suptitle("Numeric Feature Distributions", size=16)
plt.show()
```

```python
# Income vs Loan Amount
sns.scatterplot(x='Income', y='LoanAmount', hue='Default', data=df, alpha=0.4)
plt.title("Income vs Loan Amount by Default")
plt.show()
```

```python
# Credit Score vs Default
sns.kdeplot(data=df, x="CreditScore", hue="Default", fill=True)
plt.title("Credit Score Distribution by Default")
plt.show()
```

```python
# DTI Ratio vs Default
sns.boxplot(x='Default', y='DTIRatio', data=df, palette='Set2')
plt.title("DTI Ratio by Default")
plt.show()
```
