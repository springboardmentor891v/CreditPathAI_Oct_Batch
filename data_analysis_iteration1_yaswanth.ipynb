```python
# imports & load
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("Loan_Default.csv")
print(" Dataset Loaded Successfully!\n")
```

```python
# first rows, info, shape
print(df.head())
print()
print(df.info())
print()
print("Rows:", df.shape[0], "Columns:", df.shape[1])
```

```python
# missing, describe, unique counts
print(df.isnull().sum())
print()
print(df.describe())
print()
print(df.nunique())
```

```python
# correlation matrix & print
corr_matrix = df.corr(numeric_only=True)
print(corr_matrix)
```

```python
# correlation heatmap (visual)
plt.figure(figsize=(10, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Heatmap - Loan Default Dataset (Yaswanth)")
plt.show()
```

```python
# target distribution (countplot)
if "loan_default" in df.columns:
    plt.figure(figsize=(6, 4))
    sns.countplot(x="loan_default", data=df, palette="pastel")
    plt.title("Loan Default Distribution - Yaswanth Reddy")
    plt.xlabel("Loan Default (0 = No, 1 = Yes)")
    plt.ylabel("Count")
    plt.show()
else:
    print("Column 'loan_default' not found.")
```

```python
# optional relationship plot
if "loan_amount" in df.columns and "income" in df.columns:
    plt.figure(figsize=(6, 4))
    sns.scatterplot(x="income", y="loan_amount", hue="Status", data=df, palette="cool")
    plt.title("Income vs Loan Amount by Loan Status")
    plt.show()
else:
    print("Columns 'income' or 'loan_amount' not found in the dataset.")

```

```python
# Check for missing values in the dataset
print(" Checking for Missing Values:\n")
print(df.isnull().sum())
```

```python
# Correlation Matrix
plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(numeric_only=True), annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Correlation Matrix Heatmap")
plt.show()
```

```python
# Bar Chart for Loan Status (Default vs Non-Default)
if "Status" in df.columns:
    plt.figure(figsize=(6, 4))
    sns.countplot(x="Status", data=df, palette="viridis")
    plt.title("Loan Status: Default vs Non-Default")
    plt.xlabel("Status (1 = Default, 0 = Non-Default)")
    plt.ylabel("Count")
    plt.show()
else:
    print(" Column 'Status' not found in the dataset.")
```

```python
# Scatter plot to visualize income vs loan amount
if "income" in df.columns and "loan_amount" in df.columns:
    plt.figure(figsize=(6, 4))
    sns.scatterplot(x="income", y="loan_amount", hue="Status", data=df, palette="cool")
    plt.title("Income vs Loan Amount by Loan Status")
    plt.show()
else:
    print(" Columns 'income' or 'loan_amount' not found in the dataset.")
```
