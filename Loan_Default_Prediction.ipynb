```python

import pandas as pd
import sqlite3

kaggle_data = pd.read_csv('/content/Loan_Default.csv')
microsoft_data = pd.read_csv('/content/credit_risk_dataset.csv')

print("âœ… Kaggle data shape:", kaggle_data.shape)
print("âœ… Microsoft data shape:", microsoft_data.shape)

display(kaggle_data.head())
display(microsoft_data.head())

combined_data = pd.concat([kaggle_data, microsoft_data], ignore_index=True, sort=False)

conn = sqlite3.connect('/content/Loan_Default.db')
combined_data.to_sql('loan_data', conn, if_exists='replace', index=False)
conn.close()

print("\nðŸŽ¯ Data successfully ingested and stored in SQLite database!")
```
