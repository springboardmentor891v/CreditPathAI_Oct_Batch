```python
import pandas as pd

results = {
    'Model': ['Decision Tree', 'Random Forest', 'XGBoost', 'KNN', 'Logistic Regression'],
    'Train Accuracy': [1.00, 1.00, 1.00, 0.91, 0.87],
    'Train Precision': [1.00, 1.00, 1.00, 0.91, 0.92],
    'Train Recall': [1.00, 1.00, 1.00, 0.71, 0.52],
    'Train F1 Score': [1.00, 1.00, 1.00, 0.80, 0.66],
    'Train ROC-AUC': [1.00, 1.00, 1.00, 0.96, 0.86],

    'Test Accuracy': [1.00, 1.00, 0.98, 0.87, 0.87],
    'Test Precision': [1.00, 1.00, 0.98, 0.83, 0.92],
    'Test Recall': [1.00, 1.00, 1.00, 0.62, 0.53],
    'Test F1 Score': [1.00, 1.00, 0.99, 0.71, 0.68],
    'Test ROC-AUC': [1.00, 1.00, 0.99, 0.88, 0.86]
}

df_results = pd.DataFrame(results)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(18, 8))
plt.style.use('dark_background')

# Set the index
df_plot = df_results.set_index('Model')

# Generate heatmap
sns.heatmap(
    df_plot, 
    cmap='Blues', 
    annot=True,
    fmt=".4f",
    linewidths=0.5,
    linecolor='black',
    cbar=True
)

plt.title("Model Performance Comparison", fontsize=16, pad=20)
plt.tight_layout()

# Save the heatmap
plt.savefig("result_of_all_models.png", dpi=300)
plt.show()
```
