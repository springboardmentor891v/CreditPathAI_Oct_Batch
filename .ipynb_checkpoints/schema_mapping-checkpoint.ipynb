IMPORTING THE DATASET

```python
import zipfile
from tkinter import Tk
from tkinter.filedialog import askopenfilename

# Hide the main Tkinter window
Tk().withdraw()

# Open a file dialog to select the zip file
zip_path = askopenfilename(title="Select your ZIP file", filetypes=[("ZIP files", "*.zip")])

if zip_path:
    print(f"Selected file: {zip_path}")
    # Extract contents
    with zipfile.ZipFile(zip_path, 'r') as zip_ref:
        zip_ref.extractall("extracted_files")  # Folder where files will be extracted
    print("Extraction completed! Files are in the 'extracted_files' folder.")
else:
    print("No file selected.")
```
