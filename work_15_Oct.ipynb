```python
import seaborn as sns
import matplotlib.pyplot as plt

```

```python
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='Default')

plt.title('Count of Default vs Non-Default')
plt.xlabel('Default (0 = No, 1 = Yes)')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='Default', y='Income', data=df)
plt.title('Income vs Default')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Create crosstab of Default vs MaritalStatus
marital_default = pd.crosstab(df['MaritalStatus'], df['Default'])

# Plot
marital_default.plot(kind='bar', stacked=True, figsize=(7,5), color=['skyblue', 'salmon'])
plt.title('Default Distribution by Marital Status')
plt.xlabel('Marital Status')
plt.ylabel('Count')
plt.legend(title='Default', labels=['No', 'Yes'])
plt.show()
```

```python
plt.figure(figsize=(7,5))
plt.hist(df[df['Default']==0]['CreditScore'], bins=30, label='No Default', color='skyblue')
plt.hist(df[df['Default']==1]['CreditScore'],bins=30,  label='Default', color='salmon')
plt.title('Credit Score Distribution by Default Status')
plt.xlabel('Credit Score')
plt.ylabel('Count')
plt.legend()
plt.show()
```
